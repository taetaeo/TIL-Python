### 1) urlretrieve()와 urlopen()의 차이

    urlretrieve()는 바로 저장하는 방법이지만, 
    urlopen()은 데이터를 파이썬 메모리 위에 올리는 형식

```python
# 라이브러리 읽어 들이기 ---(※1)
import urllib.request 
#urllib.request처럼 점(".")으로 구분된 모듈을 지정하는 것은 urllib ㅠㅐ키지 내부에 있는 request 모듈이라는 의미이다.

# URL과 저장 경로 지정하기
url = "https://uta.pw/shodou/img/28/214.png"
savename = "test.png"
#---------------------------------------------------------------------------------------------

# 다운로드 ---(※1)
mem = urllib.request.urlopen(url).read()

# 파일로 저장하기 ---(※2)
with open(savename, mode="wb") as f: # w는 writhe(쓰기)모드 / b는 binary(이진표현)모드를 의미한다.
    f.write(mem)
    print("저장되었습니다...!")
```

### 2) 클라이언트 접속 정보 출력해보기
    ·IP주소
    ·UserAgent

```python
# IP확인 API로 접근해서 결과 출력하기
# 모듈 읽어 들이기 ---(※1)  
import urllib.request # ---urllib.request 모듈 읽어 들이기

# 데이터 읽어 들이기 ---(※2)
url = "http://api.aoikujira.com/ip/ini"
res = urllib.request.urlopen(url) # request.urlopen() 메서드를 호출하기
data = res.read() # read()메서드를 사용하여 데이터 읽기

# 바이너리를 문자열로 변환하기 ---(※3)
text = data.decode("utf-8") # decode()메서드를 사용해 바이너리를 문자열로 변환
print(text)
```

### 3) 매개변수를 추가해 요청을 전송하는 방법
    ·매개변수
    ·urllib.parse
    ·urlencode()

```python
import urllib.request
import urllib.parse
API = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
# 매개변수를 URL 인코딩합니다. ---(※1)
values = {
    "stnld": "108"  # 딕셔너리(dict) 자료형의 매개변수를 URL 인코딩을 하며, 인코딩을 위해 urllib.parse 모듈 사용
}

params = urllib.parse.urlencode(values)
#----------------------------------------------------------------------------------------------------------------

# 요청 전용 URL을 생성합니다. ---(※2)
url = API + "?" + params   # 요청전용 URL 생성 →HOW URL 끝부분에 "?"입력하고, "<KEY>=<VALUE>"형식으로 매개변수 작성
print("url=", url)

#다운로드 ---(※3)
data = urllib.request.urlopen(url).read()
text = data.decode("utf-8")
print(text)
```

### 4) 매개변수를 명령줄에서 지정하기
    ·매개변수
    ·urllib.parse
    ·urlencode()

```python
!/usr/bin/env/ python3
```

```python
# 라이버르리를 읽어 들입니다. ---(※1)
import sys
import urllib.request as req
import urllib.parse as parse

# 명령줄 매개변수 추출 ---(※2)
if len(sys.argv) <= 1:
    print("USAGE : download-forecast-argv <Region Number>")
    sys.exit()
regionNumber = sys.argv[1]

# 매개변수를 URL 인코딩합니다. ---(※3)
API = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
values = {
    'stndId' : regionNumber
}

params = parse.urlencode(values)
url = API+"?"+params
print("url = ", url)

# 다운로드 ---(※4)
data = req.urlopen(url).read()
text = data.decode("utf-8")
print(text)
```
